```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generate_data(dim=1):
    
    X = 2 * np.random.rand(100, dim)
    y = 3 + np.sum(2*X, axis=1).reshape(-1, 1) + np.random.randn(100,1)
    
    return X, y

def add_bias(X):
    
    return np.c_[np.ones((X.shape[0], 1)), X]

def get_best_param(X, y):
    
    X_t = X.T
    params = np.linalg.inv(X_t.dot(X)).dot(X_t).dot(y)
    
    return params
    
```

```python
X,y = generate_data()
```

```python
plt.scatter(X, y, c='');
```

```python
get_best_param(add_bias(X), y)
```

```python
test_X = np.array([[0], [2]])  
test_X_b = np.c_[np.ones((2, 1)), test_X]  
  
prediction = test_X_b.dot(get_best_param(add_bias(X), y))  
```

#### Gradient Descent 

```python
X,y = generate_data()
X = add_bias(X)
```

```python
X.shape
```

```python
def gradient_descent(X, y, theta, m, alpha, iterations=10):
    thetas = []
    costs = []
    for i in range(iterations):
        prediction = np.dot(X, theta)
        error = prediction - y  
        cost = (1/(2*m))*np.sum(np.dot(error, error.T))
        costs.append(cost)
        theta = theta - alpha*(1/m)*np.dot(X.T, error)
        thetas.append(theta)
        
    return thetas, costs ,theta
```

```python
thetas, costs, best_param = gradient_descent(X, y, np.random.rand(2, 1), 100, 0.01, 1000)
```
